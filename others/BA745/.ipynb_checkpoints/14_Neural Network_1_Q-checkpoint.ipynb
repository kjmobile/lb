# Neural Network 1

```python
# Import library and set the random seed for reproducibility
import tensorflow as tf
from tensorflow import keras
tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()
```

## Fashion MNIST
[MNIST Dataset Details
](https://www.tensorflow.org/api_docs/python/tf/keras/datasets/fashion_mnist/load_data)

```python
# keras.datasets.fashion_mnist.load_data() returns two objects: train, test, which can be further divided into x and y's
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

## Preprocessing

```python
train_scaled = train_input / 255.0 # normalize between 0 to 1 - this is for applying SGD
train_scaled = train_scaled.reshape(-1, 28*28) # need this because most of tensorflow classifier accept only 1darray as input.
```

```python
print(train_scaled.shape)
```

```python
from sklearn.model_selection import train_test_split

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
print(train_scaled.shape, train_target.shape)
```

```python
print(val_scaled.shape, val_target.shape)
```

## Design the Model

```python
dense = keras.layers.Dense(10, activation='softmax', input_shape=(784,))
# Used softmax as output activation since it is multiclass classification
```

```python
model = keras.Sequential(dense)
```

##  Configure and Fit the Model


```python
model.compile?
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
# configure a mdel for trainging. Syntax: model.compile(optimizer,  loss, metircs=)
# rmsprop is the default optimizer
```

```python
print(train_target[:10])
```

```python
model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
```
